```python
# import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/preprocessed_dataset.csv')

df.head(3)
```

```python
# Describe null values in dataset
for x in df.columns:
    num_nulls = df[x].isnull().sum()
    print(f"Total null values in {x}: {num_nulls}")
```

```python




# Create a heatmap to visualize the data points of the countries, year and lifeladder
pivot_df = df.pivot_table(index='region', columns='year', values='log_gdp_per_capita', aggfunc='median')
plt.figure(figsize=(10, 6))
plt.plot(pivot_df)
plt.title('log_gdp_per_capita')
plt.xlabel('Year')
plt.ylabel('Region')
plt.show()


```
